```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
plt.show()
```

```python
df["Age"].fillna(
    df["Age"].median(),
    inplace=True
)
```

```python
df["Embarked"].fillna(
    df["Embarked"].mode()[0],
    inplace=True
)
```

```python
df.drop(
    "Cabin",
    axis=1,
    inplace=True
)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["Survived"].value_counts()
```

```python
sns.countplot(
    x="Survived",
    data=df
)

plt.show()
```

```python
sns.countplot(
    x="Sex",
    hue="Survived",
    data=df
)

plt.show()
```

```python
sns.countplot(
    x="Pclass",
    hue="Survived",
    data=df
)

plt.show()
```

```python
sns.histplot(
    df["Age"],
    bins=30
)

plt.show()
```

```python
corr = df.corr(
    numeric_only=True
)
```

```python
sns.heatmap(
    corr,
    annot=True
)

plt.show()
```

```python
sns.boxplot(
    x="Survived",
    y="Age",
    data=df
)

plt.show()
```

```python
sns.boxplot(
    x="Survived",
    y="Fare",
    data=df
)

plt.show()
```
